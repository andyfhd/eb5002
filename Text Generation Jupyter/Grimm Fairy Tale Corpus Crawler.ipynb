```python
taledict = dict(zip(title,url))
Grimm_text=''
counter=0
for title, url in taledict.items():
    Grimm_text = Grimm_text+'<{}>\n\n'.format(title)
    counter+=1
    print(counter,title)
    for line in urlopen(url):
        Grimm_text = Grimm_text+str(line.decode("utf-8"))
    #if counter>2:break
text_file = open("data/Grimm_text.txt", "w", encoding="utf-8") #encoding needs to be utf-8
text_file.write(Grimm_text)
text_file.close()
```
